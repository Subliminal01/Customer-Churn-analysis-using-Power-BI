```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/E Commerce Dataset.csv')
```

```python
df.head()
```

```python
# Drop CustomerID
df.drop(columns=['CustomerID'], inplace=True)
```

```python
#columns_with_missing_values =
df.isna().any()
```

```python
df.isnull().sum()
```

```python
for cols in df.columns:
  if df[cols].isnull().sum()>0:
    df[cols].fillna(df[cols].median(), inplace=True)
```

```python
df.head()
```

```python
#no null values now
df.isna().any()
```

```python
sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()
```

```python
df.dtypes
```

```python
#categorical variables:
categorical_variables=df.select_dtypes(include=['object']).columns
categorical_variables
```

```python
for cols in categorical_variables:
  sns.histplot(data=df, x=cols, hue='Churn', multiple='stack', shrink=0.8)
  plt.title(f"Churn by {cols} Type")
  plt.xticks(rotation=45)
  plt.show()
```

```python
sns.barplot(data=df, x='Churn', y='Tenure', estimator='mean')
plt.title('Churn')
```

```python
sns.violinplot(data=df, x='Churn', y='Tenure', split=True)
plt.title('Tenure Distribution by Churn')
```

```python
df.to_csv('/content/E Commerce Dataset processed.csv')
```

```python
df_y=df['Churn']
df.drop(columns=['Churn'], inplace=True)
```

```python
#numerical data type columns
numerical_data_types=df.select_dtypes(include=['float64', 'int64']).columns
numerical_data_types
```

```python
for cols in numerical_data_types:
  sns.boxplot(data=df, x=df_y, y=cols)
  plt.xticks([0, 1], ['Retained (0)', 'Churned (1)'])
  plt.title(f"{cols} by Churn Status chart")
  plt.show()
```

```python
#complaint
import seaborn as sns
confusion_matrix = pd.crosstab(df_y, df['Complain'])
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Churn')
plt.ylabel('Complain')
```
